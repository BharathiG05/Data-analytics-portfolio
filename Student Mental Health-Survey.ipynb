### RENAME COLUMNS

```python
df.rename(columns={'Choose your gender':'Gender','What is your course?':'Course','Your current year of Study':'Year',
                   'What is your CGPA?':'CGPA','Do you have Depression?':'Depression','Do you have Anxiety?':'Anxiety',
                   'Do you have Panic attack?':'Panic_Attack','Did you seek any specialist for a treatment?':'Specialist_Treatment'},inplace=True)
df.columns
```

```python
df.rename(columns={'Marital status':'Marital_status'},inplace=True)
df.columns
```

```python
df.head(10)
```

### STRIP COLUMNS

```python
str_cols = ['Gender', 'Course', 'Year', 'CGPA', 
            'Marital_status', 'Depression', 
            'Anxiety', 'Panic_Attack', 'Specialist_Treatment']
for col in str_cols:
    df[col] = df[col].str.strip()
print('whitespace removed')
```

### HANDLE MISSING VALUES

```python
print("Missing values:\n", df.isnull().sum())

# Fill missing Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)
print("\nMissing values after:\n", df.isnull().sum())
```

### FIX YEAR COLUMN

```python
df['Year'] = df['Year'].str.title()
print(df['Year'].unique())
```

### FIX DATA TYPES

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age'] = df['Age'].astype(int)
print(df.dtypes)
```

### REMOVE DUPLICATES


```python

print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

### RESET INDEX

```python
df.reset_index(drop=True, inplace=True)
print("Data Cleaning Complete")
print("Final shape:", df.shape)
df.head()
```

## 2. DATA TRANSFORMATION

### ENCODE YES/NO COLUMNS (Text → Numbers) 

```python
yes_no_map = {'Yes': 1, 'No': 0}

df['Anxiety'] = df['Anxiety'].map(yes_no_map)
df['Panic_Attack'] = df['Panic_Attack'].map(yes_no_map)
df['Specialist_Treatment'] = df['Specialist_Treatment'].map(yes_no_map)

print(df[['Depression', 'Anxiety', 'Panic_Attack']].head())
```

### ENCODE GENDER 

```python
gender_map = {'Male': 'M', 'Female': 'F'}
df['Gender'] = df['Gender'].map(gender_map)
print(df['Gender'].value_counts())
```

### ENCODE YEAR OF STUDY 


```python
year_map = {
    'Year 1': 1,
    'Year 2': 2,
    'Year 3': 3,
    'Year 4': 4
}
df['Year_Num'] = df['Year'].map(year_map)
print(df[['Year', 'Year_Num']].head())
```

### ENCODE CGPA 

```python

cgpa_map = {
    '0 - 1.99': 8,
    '2.00 - 2.49': 9,
    '2.50 - 2.99': 6,
    '3.00 - 3.49': 7,
    '3.50 - 4.00': 8
}
df['CGPA_Score'] = df['CGPA'].map(cgpa_map)
print(df[['CGPA', 'CGPA_Score']].head())
```

```python
df
```

### EXTRACT DATE PARTS FROM TIMESTAMP

```python
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day
df['Hour'] = df['Timestamp'].dt.hour

print(df[['Timestamp', 'Month', 'Day', 'Hour']].head())
```

### NORMALIZE AGE 

```python
df['Age_Norm'] = (df['Age'] - df['Age'].min()) / \
                 (df['Age'].max() - df['Age'].min())

print(df[['Age', 'Age_Norm']].head())
```

# 3.DATA ANALYSIS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### DEPRESSION COUNT (Bar Chart)

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Depression', palette='viridis')
plt.title('Depression Distribution')
plt.xlabel('Depression (0=No, 1=Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

OBSERVATION:
1.The bar chart compares the number of students who reported experiencing depression with those who did not.
2.The 'No' category has a higher count than the 'Yes' category, indicating that most students in the survey do not report depression.
3.However, a notable proportion of students reported experiencing depression, showing that mental health is a significant concern among students.
4.The results emphasize the importance of providing mental health awareness, counseling services, and early support programs in educational institutions

### GENDER DISTRIBUTION (Pie Chart)

```python
plt.figure(figsize=(7, 7))
df['Gender'].value_counts().plot.pie(
    labels=['Female', 'Male'],
    autopct='%1.1f%%', 
    startangle=90,
    colors=['pink', 'steelblue'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

OBSERVATION:
1.The pie chart shows that 74.3% of the survey participants are female, while 25.7% are male. This indicates that the dataset is predominantly composed of female students, meaning the survey responses mainly reflect the mental health experiences of female participants. 
2.Therefore, the findings may be more representative of female students than male students.

### CGPA VS DEPRESSION (Bar Chart) 

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='CGPA_Score', hue='Depression', palette='Set1')
plt.title('CGPA vs Depression')
plt.xlabel('CGPA Range')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

OBSERVATION:
1.The chart compares the number of students with and without depression across different CGPA ranges.
2.Students with both high and low CGPA reported depression, indicating that depression is not limited to a specific academic performance level.
3.In most CGPA ranges, the number of students without depression is higher than those with depression.
4.The chart suggests that academic performance alone does not determine a student's mental health, as depression can affect students regardless of their CGPA.
5.These findings highlight the importance of providing mental health support to all students, irrespective of their academic achievements.

### ANXIETY VS PANIC ATTACK (Count Plot)

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Anxiety', hue='Panic_Attack', palette='coolwarm')
plt.title('Anxiety vs Panic Attack')
plt.xlabel('Anxiety (0=No, 1=Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

OBSERVATION:
1.The chart compares the occurrence of panic attacks among students with and without anxiety.
2.Students who reported anxiety show a higher number of panic attacks compared to students without anxiety.
3.Students without anxiety generally report fewer panic attacks, indicating a lower prevalence.
4.This suggests a strong association between anxiety and panic attacks, where students experiencing anxiety are more likely to experience panic attacks.
5.The findings highlight the importance of early identification and support for students with anxiety to help reduce the risk of panic attacks.

### AGE DISTRIBUTION (Histogram)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='orange')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

OBSERVATION:
1.Most students are concentrated within a young adult age group, indicating that the survey primarily represents college and university students.
2.The frequency of students decreases as the age moves away from the central age range, showing that fewer respondents belong to older or younger age groups.
3.The distribution suggests that the dataset mainly consists of students of a similar age, making it appropriate for analyzing mental health among college students

### CORRELATION HEATMAP

```python
# cr=df.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(),
            annot=True, cmap='Blues', fmt='.2f')
plt.title('Correlation Heatmap - Blues')
plt.tight_layout()
plt.show()
```

OBSERVATION:
1.Tha correlation heatmap illustrates the strength and direction of relationships between the numerical variables in the student mental health dataset.
2.Positive correlation values indicate that two variables tend to increase together, while negative correlation values indicate that one variable tends to decrease as the other increases.
3.The heatmap shows that depression, anxiety, and panic attacks are positively correlated, suggesting that these mental health conditions are closely associated.
4.Variables with correlation values close to 0 have little or no linear relationship.
5.Overall, the heatmap helps identify the variables that are most strongly related, providing valuable insights into the factors associated with students' mental health.
